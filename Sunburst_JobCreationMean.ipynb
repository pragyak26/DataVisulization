```python
sb = list()
```

```python
for state in unique_states:
  # print(state, df.loc[df['State'] == state, 'Data.Job Destruction.Rate'].mean())
  m=(df.loc[df['State'] == state, 'Data.Job Creation.Rate'].mean())
  sb.append(m)

print(sb)
```

```python
test_list=[unique_states, sb]
```

```python
newData = pd.DataFrame(test_list, index=['state', 'mean']).T
newData
```

```python
def viz_meanOfState(data):
  
  fig = px.sunburst(data, path=['state','mean'], values='mean',
                    color='mean', hover_data=['mean'],
                    color_continuous_scale='RdBu',)
                                        
  fig.show()
```

```python
viz_meanOfState(newData)
```
